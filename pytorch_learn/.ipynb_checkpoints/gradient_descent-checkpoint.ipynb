```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = [1.0,2.0,3.0]
y_data = [2.0,4.0,6.0]
cost_list=[]
w = 1.0
```

```python
def forward(x):
    return x*w
```

```python
def gradient_descent(xs,ys):
    grad = 0
    for x,y in zip(xs,ys):
        grad += 2 * x * (x * w - y)
    return grad/len(xs)
```

```python
def cost(xs,ys):
    cost = 0
    for x,y in zip(xs,ys):
        y_pred = forward(x)
        cost += (y_pred - y)**2
    return cost/len(xs)
    
```

```python
print("Predict (before training)",4,forward(4))

for epoch in range(100):
    # 计算损失值
    cost_val = cost(x_data,y_data)
    cost_list.append(cost_val)
    # 梯度值计算
    grad_val = gradient_descent(x_data,y_data)
    # 更新权重
    w -= 0.01 * grad_val
    
    print("Epoch:",epoch,"w=",w,"loss=",cost_val)

print("Predict (after training)",4,forward(4))
```

```python
plt.plot(range(100),cost_list)
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.show
```

```python
# 随机梯度下降
# 计算梯度时不使用样本整体的代价函数求偏导，而是使用每一个样本的损失函数求偏导
# 随机梯度下降考量了局部样本的损失函数，可以很好地解决权重更新时陷入鞍点导致迭代停止的问题
def stotastic_grad(x,y):
    return  2 * x * (x * w - y)
```

```python
def loss(x,y):
    y_pred = forward(x)
    return (y_pred-y)**2
```

```python
loss_list=[]
w=1.0

print("Predict (before training)",4,forward(4))

for epoch in range(100):
    loss_val = 0
    for x,y in zip(x_data,y_data):
        # 计算损失值
        loss_val +=loss(x,y)
        # 梯度值计算
        grad_val = stotastic_grad(x,y)
        # 更新权重
        w -= 0.01 * grad_val
    loss_list.append(loss_val/3)
    print("Epoch:",epoch,"w=",w,"loss=",loss_val)

print("Predict (after training)",4,forward(4))
```

```python
print(loss_list)
```

```python
plt.plot(range(100),loss_list)
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.show
```
